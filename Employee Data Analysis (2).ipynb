```python
import pandas as pd
df=pd.read_csv(r"C:\Users\ELCOT\Desktop\employee_performance_workload_attrition.csv")
```

```python
attrition_rate = (
    df['attrition'].value_counts(normalize=True) * 100
)

print("\nOverall Attrition Rate (%)")
print(attrition_rate)
```

```python
dept_attrition =pd.crosstab(
    df['department'],
    df['attrition'],
    normalize='index'
) * 100

print("\nAttrition by Department (%)")
print(dept_attrition)
```

```python
experience_attrition = pd.crosstab(
    df['role_level'],
    df['attrition'],
    normalize='index'
) * 100

print("\nAttrition by Role Level (%)")
print(experience_attrition)
```

```python
numeric_cols = [
    'monthly_salary',
    'avg_weekly_hours',
    'projects_handled',
    'performance_rating',
    'absences_days',
    'job_satisfaction'
]
print(numeric_cols)
```

```python
df['attrition_flag'] = df['attrition'].map(
    {'Yes': 1, 'No': 0}
)

correlation = df[numeric_cols + ['attrition_flag']].corr()

print("\nCorrelation with Attrition:")
print(
    correlation['attrition_flag']
    .sort_values(ascending=False)
)
```

```python
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\ELCOT\Desktop\employee_performance_workload_attrition.csv")
```

```python
plt.figure(figsize=(6,6))
df['attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title("Employee Attrition Distribution")
plt.ylabel("")
plt.show()
```

```python
dept_yes = (
    df[df['attrition'] == 'Yes']
    .groupby('department')
    .size()
)

plt.figure(figsize=(8,5))
dept_yes.plot(kind='bar')
plt.title("Employees Leaving by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```
